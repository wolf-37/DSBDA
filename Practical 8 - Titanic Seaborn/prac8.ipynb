```python
# import and load dataset
import seaborn as sns
df= sns.load_dataset('titanic')
df
```

```python
# getting relevent columns
data = df[['survived','class','sex','age','fare']]
data
```

## A JointPlot in Seaborn is a combination of a scatter plot (or other plots) for two variables, along with their individual distributions (histograms or KDE) on the margins.

```python
sns.jointplot(x='age',y='fare',data=data)
```

```python
sns.jointplot(x='age', y='fare', data=data, hue='survived')
```

```python
sns.jointplot(x='age', y='fare', data=data, hue='class')
```

## A PairPlot plots pairwise relationships in a dataset â€” all combinations of numeric variables against each other, along with histograms/KDE on the diagonal.

```python
sns.pairplot(data, hue='sex')
```

## A CountPlot is a bar chart that shows the number of occurrences of each category in a categorical variable.

```python
sns.countplot(x=data['sex'])
```

```python
sns.countplot(x=data['class'])
```

## A BarPlot shows the relationship between a categorical variable and a numeric variable using rectangular

```python
sns.barplot(x='sex',y='survived',data=data)
```

```python
sns.barplot(x='sex',y='survived',hue='class',data=data)
```

## A HistPlot (Histogram) visualizes the distribution of a single continuous variable by dividing the range into bins and counting how many values fall into each bin.

```python
sns.histplot(data['fare'])
```

## A KDEPlot (Kernel Density Estimate) is a smoothed version of a histogram, showing the estimated probability density function (PDF) of a continuous variable.

```python
sns.kdeplot(data['fare'])
```
