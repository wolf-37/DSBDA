```python
# importing required libraries
import pandas as pd
import numpy as np 
```

```python
# reading data as Dataframe
df = pd.read_csv("BostonHousing.csv")
df
```

```python
# Get NA value 
df.isna()
```

```python
# Get NA value count per Column 
df.isna().sum()
```

```python
# Absolute value fill NA Value
df.fillna(0)
df
```

```python
# Forward fill NA Value
df.fillna(method='ffill')
df
```

```python
# Backward fill NA Value
df.fillna(method='bfill')
df
```

```python
# mean fill NA Value
df['rm'].fillna(df['rm'].mean())
df
```

```python
# median fill NA Value
df['rm'].fillna(df['rm'].median())
df
```

```python
# Drop NA Value
df.dropna(inplace = True)
```

```python
# input
inp = df.drop(columns=["medv"])
inp[75:80]
```

```python
# output
out = df["medv"]
out
```

```python
# to split data into train and test sets
from sklearn.model_selection import train_test_split as tts
```

```python
trainx, testx, trainy, testy = tts(inp, out, test_size=0.2, random_state=36)
```

```python
trainx
```

```python
trainy
```

```python
testx
```

```python
testy
```

```python
# Importing classifier LinearRegression
from sklearn.linear_model import LinearRegression as LR
```

```python
# model building
model = LR()
model.fit(trainx, trainy)
```

```python
# m1, m2, ... values
model.coef_
```

```python
# c value
model.intercept_
```

```python
model.score(testx,testy)
```

```python
# Test new data
new = np.array( [0.05646,	0.0,	12.83,	0,	0.437,	6.232,	53.7,	5.0141,	5,	398,	18.7,	386.40,	12.34])
new
```

```python
# Predict value
model.predict([new])
```

```python
# Predict xtest
predy = model.predict(testx)
predy
```

```python
# Create testy(actual) predy(predicted by model) residue( abs(testy-predy) )
r = pd.DataFrame({"Actual" : testy, "Predicted" : predy, "Residue" : abs(testy-predy)})
r
```

```python
# Mean Absolute Error Manually
m = ( sum(r['Residue']) )/( len(r['Residue']) )
m
```

```python
# Mean Absolute Error using Library
from sklearn.metrics import mean_absolute_error as mae
m = mae(r['Actual'],r['Predicted'])
m
```
