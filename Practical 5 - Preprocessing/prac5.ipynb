```python
# importing required libraries
import pandas as pd
import numpy as np 
```

```python
# reading data as Dataframe
df = pd.read_csv("SocialNetworkAds.csv")
df
```

```python
# Get NA value 
df.isna()
```

```python
# Get NA value count per Column 
df.isna().sum()
```

```python
# Drop NA Value
df.dropna(inplace = True)
```

```python
# input
inp = df[['Age','EstimatedSalary']]
inp
```

```python
# output
out = df["Purchased"]
out
```

```python
import seaborn as sns
sns.countplot(x=out)
```

```python
# scales and transforms features to a fixed range — usually between 0 and 1
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
inp_scaled = scaler.fit_transform(inp)
inp_scaled
```

```python
# to split data into train and test sets
from sklearn.model_selection import train_test_split as tts

trainx, testx, trainy, testy = tts(inp_scaled, out, test_size=0.2, random_state=36)
```

```python
trainx
```

```python
trainy
```

```python
testx
```

```python
testy
```

```python
# Importing classifier LinearRegression
from sklearn.linear_model import LogisticRegression as LR
```

```python
# model building
model = LR()
model.fit(trainx, trainy)
```

```python
# m1, m2, ... values
model.coef_
```

```python
# c value
model.intercept_
```

```python
model.score(testx,testy)
```

```python
# Predict xtest
predy = model.predict(testx)
predy
```

```python
# Create testy(actual) predy(predicted by model) )
r = pd.DataFrame({"Actual" : testy, "Predicted" : predy})
r
```

```python
# Mean Absolute Error using Library
from sklearn.metrics import mean_absolute_error as mae
m = mae(r['Actual'],r['Predicted'])
m
```

```python
# get Confusion matrix
from sklearn.metrics import confusion_matrix

confusion_matrix(r['Actual'],r['Predicted'])
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

cm = confusion_matrix(r['Actual'],r['Predicted'])
disp = ConfusionMatrixDisplay(confusion_matrix = cm)

disp.plot(cmap='Blues')
plt.show()
```
